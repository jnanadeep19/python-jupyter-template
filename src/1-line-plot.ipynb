### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 11)
y = x ** 2

plt.figure(figsize=(8, 5))
plt.plot(x, y, color='royalblue', linewidth=2, marker='o')
plt.title('Square of X Values', fontsize=14)
plt.xlabel('X Value', fontsize=12)
plt.ylabel('Y = XÂ²', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```
